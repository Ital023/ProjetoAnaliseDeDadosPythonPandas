```python
import pandas as pd
import matplotlib.pyplot as plt

#Lendo a base de dados
Aracaju = pd.read_excel(r"C:\Users\ItaloMiranda\Documents\GIT\DIO\ProjetoPandasEexcel\BaseDeDados\Aracaju.xlsx")
Fortaleza = pd.read_excel(r"C:\Users\ItaloMiranda\Documents\GIT\DIO\ProjetoPandasEexcel\BaseDeDados\Fortaleza.xlsx")
Natal = pd.read_excel(r"C:\Users\ItaloMiranda\Documents\GIT\DIO\ProjetoPandasEexcel\BaseDeDados\Natal.xlsx")
Recife = pd.read_excel(r"C:\Users\ItaloMiranda\Documents\GIT\DIO\ProjetoPandasEexcel\BaseDeDados\Recife.xlsx")
Salvador = pd.read_excel(r"C:\Users\ItaloMiranda\Documents\GIT\DIO\ProjetoPandasEexcel\BaseDeDados\Salvador.xlsx")

#Concatenar todas as planilhas
df = pd.concat([Aracaju,Fortaleza,Natal,Recife,Salvador])
df.dtypes
```

```python
#Organizando as datas no dataframe
df["Dia_Venda"] = df ["Data"].dt.day
df["Mes_Venda"] = df ["Data"].dt.month
df["Ano_Venda"] = df ["Data"].dt.year

display (df)
```

```python
#Calculando a receita gerada
df["Receita"] = df["Vendas"] * df ["Qtde"]

display(df)
```

```python
#Verificando se tem valor nulo
df.isnull().sum()
```

```python
#Analisando os dados das 10 maiores receitas
df.nlargest(10,"Receita")
```

```python
#Analisando os dados das 10 menores receitas
df.nsmallest(10,"Receita")
```

```python
#Grafico Total das melhores cidades que mais venderam quantidade
df.groupby("Cidade")["Qtde"].sum().sort_values(ascending = True).plot.barh(title = "Cidades que mais venderam quantidade")
plt.xlabel ("Quantidade")
plt.ylabel("Cidades")
```

```python
#Verificando a validade da informacao
df.groupby("Cidade")["Qtde"].sum().sort_values(ascending = False)
```

```python
#Grafico total das melhores cidades e suas receitas
df.groupby("Cidade")["Receita"].sum().sort_values(ascending = True).plot.barh(title = "Melhores receitas e suas cidades",color = "green")
plt.xlabel ("Receita")
plt.ylabel("Cidades")
```
